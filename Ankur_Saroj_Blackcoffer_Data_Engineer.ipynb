```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
#reading input.xlsx file for the urls

df=pd.read_excel(r"D:\practice python\Web scraping\Blackcoffer_assign\Data_Engineer\Input.xlsx")
```

```python
df.head()
```

```python
#creating .txt file for all the text scrabbed from the websites in df['URL']
```

```python
i=1
for x in df['URL']:
    try:
        re=rq.get(x)
        soup=bs(re.text)
        with open(f"D:/practice python/Web scraping/Blackcoffer_assign/Data_Engineer/text_files/file_{i}.txt",'w') as file:
            try:
                file.write(soup.text)
                i=i+1
            except:
                print(f"\n file_{i}.txt not created.........some error has occured.......\n")
                i=i+1
    except:
        print(f"\n Broke for file_{i}")
        i=i+1
```

```python
#Sentiment Analysis using nltk

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.probability import FreqDist
from textstat import flesch_reading_ease, syllable_count
# fog_index
```

```python
i=0
res={"Positive_score":[],"Negative_score":[],"Polarity_score":[],"Subjectivity_score":[],"Avg_sentence_length":[],"Percentage_complex_words":[],"Fog_index_score":[],"Avg_word_per_sentence":[],"Complex_word_count":[],"Word_count":[],"syllables_per_word":[],"Personal_Pronouns":[],"Avg_word_length":[]}
while i<len(df['URL']):
    #reading the .txt file which we have created above....
    i=i+1
    file=open(f"D:/practice python/Web scraping/Blackcoffer_assign/Data_Engineer/text_files/file_{i}.txt",'r')
    main=file.read()
    
    if main!='':
        sentences = sent_tokenize(main)
        words = word_tokenize(main)

        analyzer = SentimentIntensityAnalyzer()
        scores = analyzer.polarity_scores(main)
        positive_score = scores['pos']
        negative_score = scores['neg']
        polarity_score = scores['compound']
        subjectivity_score = 1 - scores['neu']


        avg_sentence_length = len(words) / len(sentences)
        complex_words = [word for word in words if syllable_count(word) > 2]
        percentage_complex_words = len(complex_words) / len(words) * 100
        # fog_index_score = fog_index(main)
        avg_words_per_sentence = len(words) / len(sentences)
        complex_word_count = len(complex_words)
        fog_index_score=0.4*(avg_words_per_sentence+((len([word for word in words if syllable_count(word) > 3])/len(words))*100))
        word_count = len(words)
        syllables_per_word = syllable_count(main) / word_count


        freq_dist = FreqDist(words)
        personal_pronouns = freq_dist['I'] + freq_dist['me'] + freq_dist['we'] + freq_dist['us'] + freq_dist['you'] + freq_dist['he'] + freq_dist['him'] + freq_dist['she'] + freq_dist['her'] + freq_dist['it'] + freq_dist['they'] + freq_dist['them']
        avg_word_length = sum(len(word) for word in words) / len(words)

        #appending all the find-out results into dictionary 
        
        res['Positive_score'].append(positive_score)
        res['Negative_score'].append(negative_score)
        res['Polarity_score'].append(polarity_score)
        res['Subjectivity_score'].append(round(subjectivity_score,3))
        res['Avg_sentence_length'].append(round(avg_sentence_length,3))
        # res['Complex_words'].append(complex_words)
        res['Percentage_complex_words'].append(round(percentage_complex_words,3))
        res['Fog_index_score'].append(fog_index_score)
        res['Avg_word_per_sentence'].append(round(avg_words_per_sentence,3))
        res['Complex_word_count'].append(complex_word_count)
        res['Word_count'].append(word_count)
        res['syllables_per_word'].append(round(syllables_per_word,3))
        res['Personal_Pronouns'].append(personal_pronouns)
        res['Avg_word_length'].append(round(avg_word_length,3))
    else:
        #if file_{i} is null, then initialize all the columns with 0;
        res['Positive_score'].append(0)
        res['Negative_score'].append(0)
        res['Polarity_score'].append(0)
        res['Subjectivity_score'].append(0)
        res['Avg_sentence_length'].append(0)
        # res['Complex_words'].append(complex_words)
        res['Percentage_complex_words'].append(0)
        res['Fog_index_score'].append(0)
        res['Avg_word_per_sentence'].append(0)
        res['Complex_word_count'].append(0)
        res['Word_count'].append(0)
        res['syllables_per_word'].append(0)
        res['Personal_Pronouns'].append(0)
        res['Avg_word_length'].append(0)
        
    file.close()
    
```

```python
if(len(df['URL'])==len(res['Positive_score'])):
    print('Well done!')
```

```python
#now your data is ready to change into DataFrame......
#DataFrame of the findout result
df1=pd.DataFrame(res)
```
